```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Year': [2015, 2016, 2017, 2018, 2019],
        'Sales': [100, 150, 200, 180, 220],
        'Expenses': [80, 100, 120, 100, 130]}

df = pd.DataFrame(data)

# Line plot
df.plot(x='Year', y='Sales', kind='line')
plt.title('Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales (in units)')
plt.show()

# Bar plot
df.plot(x='Year', y='Sales', kind='bar')
plt.title('Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales (in units)')
plt.show()

# Scatter plot
df.plot(x='Sales', y='Expenses', kind='scatter')
plt.title('Expenses vs Sales')
plt.xlabel('Sales (in units)')
plt.ylabel('Expenses (in units)')
plt.show()

# Histogram
df['Sales'].plot(kind='hist')
plt.title('Sales Distribution')
plt.xlabel('Sales (in units)')
plt.ylabel('Frequency')
plt.show()

```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 2, 3], label='Series 1')
plt.plot([1, 2, 3, 4], [4, 3, 1, 3], label='Series 2')

plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Plot Title')

plt.text(3, 4, 'Example Annotation', fontsize=12, color='red')
plt.annotate('Reversal', xy=(3, 2), xytext=(2, 1.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate('', xy=(3, 1), xytext=(2.4, 1.5),
             arrowprops=dict(facecolor='red', arrowstyle='->'))

plt.show()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as patches

plt.plot([1, 2, 3, 4], [1, 4, 2, 3], label='Series 1')
plt.plot([1, 2, 3, 4], [4, 3, 1, 3], label='Series 2')

plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Plot Title')

plt.text(3, 4, 'Example Annotation', fontsize=12, color='red')
plt.annotate('Reversal', xy=(3, 2), xytext=(2, 1.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate('', xy=(3, 1), xytext=(2, 1),
             arrowprops=dict(facecolor='red', arrowstyle='->'))

# Create circle annotation
circle = patches.Circle((2, 4), 0.4, facecolor='blue', alpha=0.3)
plt.gca().add_patch(circle)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

plt.plot([1, 2, 3, 4], [1, 4, 2, 3], label='Series 1')
plt.plot([1, 2, 3, 4], [4, 3, 1, 3], label='Series 2')

plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Plot Title')

plt.text(3, 4, 'Example Annotation', fontsize=12, color='red')
plt.annotate('Reversal', xy=(3, 2), xytext=(2, 1.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate('', xy=(3, 1), xytext=(2.4, 1.5),
             arrowprops=dict(facecolor='red', arrowstyle='->'))

# Create circle
circle = patches.Circle((2, 4), 0.4, facecolor='blue', alpha=0.3)
plt.gca().add_patch(circle)

# Create rectangle
rect = patches.Rectangle((1, 4.5), 3, 0.7, facecolor='red', alpha=0.3)
plt.gca().add_patch(rect)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.lines as lines

plt.plot([1, 2, 3, 4], [1, 4, 2, 3], label='Series 1')
plt.plot([1, 2, 3, 4], [4, 3, 1, 3], label='Series 2')

plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Plot Title')

plt.text(3, 4, 'Example Annotation', fontsize=12, color='red')
plt.annotate('Reversal', xy=(3, 2), xytext=(2, 1.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate('', xy=(3, 1), xytext=(2.4, 1.5),
             arrowprops=dict(facecolor='red', arrowstyle='->'))

# Circle
circle = patches.Circle((2, 4), 0.4, facecolor='blue', alpha=0.3)
plt.gca().add_patch(circle)

# Rectangle
rect = patches.Rectangle((1, 4.5), 3, 0.7, facecolor='red', alpha=0.3)
plt.gca().add_patch(rect)

# Line 1
line1 = lines.Line2D([2.5, 3.5], [1, 1], linewidth=2, color='green')
plt.gca().add_line(line1)

# Line 2
line2 = lines.Line2D([1.5, 2.5], [4, 4], linewidth=2, color='purple')
plt.gca().add_line(line2)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 15, 12, 17, 9]
y2 = [5, 8, 6, 12, 10]
y3 = [7, 10, 9, 13, 11]

# Create a figure with 2 rows and 2 columns of subplots

# First subplot
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Subplot 1')

# Second subplot
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Subplot 2')

# Third subplot
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Subplot 3')

# Optional: Adjust spacing between subplots
plt.tight_layout()

# Display the subplots
plt.show()
```
